```python
# special IPython command to prepare the notebook for matplotlib
%matplotlib inline 

import numpy as np
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt
```

### Normal Distribution Discretized

```python
mean = 100.0 # Make sure to use float
sd = 20.0 # Make sure to use float
low = mean - 3 * sd # low end of x axis
high = mean + 3 * sd # high end of x axis

dx = 2 # Length of interval

# Comb points along x axis
x = np.arange(low, high, dx)

# Compute y values: pdf at each value of x
y = (1/(sd * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean)/sd) ** 2)

# Plot the function
plt.figure(figsize = (16, 8))
plt.stem(x, y, markerfmt = ' ') # This draws the intervals
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.title('Discretized Normal Probability Distribution')

area = np.sum(dx * y)
print 'Probability Sum = %1.4f' % area
```
