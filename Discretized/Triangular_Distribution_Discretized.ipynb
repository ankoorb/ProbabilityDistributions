```python
# special IPython command to prepare the notebook for matplotlib
%matplotlib inline 

import numpy as np
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt
```

### Triangular Distribution Discretized

```python
min_x = 0.2
mean_x = 0.4
max_x = 0.75
x_array = np.random.triangular(min_x, mean_x, max_x, size = 100000)

val = np.histogram(x_array, bins = 20)
y = [float(i)/np.sum(val[0]) for i in val[0]]

# Binwidth issue
x_x = val[1]
x = []
for i in range(len(x_x) - 1):
    temp = (x_x[i] + x_x[i+1])/2
    x.append(temp)

# Plot triangular distribution
plt.figure(figsize = (16, 8))
plt.stem(x, y, markerfmt = ' ') # This draws the intervals
plt.xlabel('$x$')
plt.ylabel('$p(x)$')
plt.title('Discretized Triangular Probability Distribution')

print 'Probability Sum = %1.4f' % np.sum(y)
```
